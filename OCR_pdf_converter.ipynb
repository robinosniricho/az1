```python
import os
import re
import subprocess
import gdown
from google.colab import files

# Step 1: Ask for the Google Drive link of your PDF (make sure it's shareable with "Anyone with the link")
gdrive_link = input("Enter the Google Drive link of your PDF: ")

# Step 2: Extract the file ID from the Google Drive link using regex
file_id_match = re.search(r"/d/([a-zA-Z0-9_-]+)", gdrive_link)
if not file_id_match:
    raise ValueError("Invalid Google Drive link! Please provide a correct shareable link.")
file_id = file_id_match.group(1)

# Step 3: Define the local filename for the downloaded PDF
downloaded_pdf = "book.pdf"

# Download the PDF from Google Drive using gdown
print("Downloading PDF from Google Drive...")
gdown.download(f"https://drive.google.com/uc?id={file_id}", downloaded_pdf, quiet=False)

# Step 4: Define the output filename for the OCR-processed PDF
output_pdf = "converted_book.pdf"

# Step 5: Run OCR on the downloaded PDF using ocrmypdf with --force-ocr to force processing
print("Applying OCR to the PDF. This may take several minutes...")
subprocess.run(["ocrmypdf", "--force-ocr", downloaded_pdf, output_pdf], check=True)

# Step 6: Check if the converted PDF exists and then download it
if os.path.exists(output_pdf):
    print(f"OCR process completed! Your converted PDF is saved as: {output_pdf}")
    files.download(output_pdf)
else:
    raise FileNotFoundError("OCR failed! The converted PDF was not found.")
```
