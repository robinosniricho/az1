```python
import cv2
import os

def extract_frames(video_path, output_folder, frame_interval=5):
    os.makedirs(output_folder, exist_ok=True)
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    saved_count = 0

    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        if frame_count % frame_interval == 0:
            frame_path = os.path.join(output_folder, f"frame_{saved_count}.jpg")
            cv2.imwrite(frame_path, frame)
            saved_count += 1

        frame_count += 1

    cap.release()
    print(f"Extracted {saved_count} frames.")

# Example usage
extract_frames("video.mp4.webm", "frames", frame_interval=5)
```

```python
from paddleocr import PaddleOCR
import os

# Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang="en")

def extract_text_from_images(input_folder, output_file="output.txt"):
    with open(output_file, "w") as f:
        for image_name in sorted(os.listdir(input_folder)):
            image_path = os.path.join(input_folder, image_name)
            result = ocr.ocr(image_path, cls=True)

            # ✅ Check if OCR detected any text before processing
            if result and result[0]:
                f.write(f"\nText from {image_name}:\n")
                for line in result[0]:  # Fix: Access first element safely
                    if line and len(line) > 1:  # Ensure valid structure
                        f.write(line[1][0] + "\n")  # Extract detected text
            else:
                print(f"Warning: No text detected in {image_name}")

    print(f"Text extraction completed. Results saved in {output_file}")

# Example usage
extract_text_from_images("frames")
```

```python
import os
from paddleocr import PaddleOCR
import concurrent.futures

def extract_text(image_path):
    """
    Process a single image with PaddleOCR and return its detected text.
    Each process creates its own PaddleOCR instance.
    """
    ocr = PaddleOCR(use_angle_cls=True, lang="en")  # Initialize OCR for this worker.
    result = ocr.ocr(image_path, cls=True)
    extracted_text = []
    # Check if any text is detected.
    if result and result[0]:
        for line in result[0]:
            # Make sure the structure is valid before extracting the text.
            if line and len(line) > 1:
                extracted_text.append(line[1][0])
    return (os.path.basename(image_path), extracted_text)

def main():
    image_folder = "frames"   # Folder containing images for OCR.
    output_file = "ocr_output_parallel.txt"

    # Get a sorted list of all .jpg files in the folder.
    image_files = sorted([os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.lower().endswith('.jpg')])

    results = []
    # Process the images in parallel.
    with concurrent.futures.ProcessPoolExecutor() as executor:
        futures = {executor.submit(extract_text, image_file): image_file for image_file in image_files}
        for future in concurrent.futures.as_completed(futures):
            image_name, text = future.result()
            results.append((image_name, text))

    # Save the OCR results to an output text file.
    with open(output_file, "w", encoding="utf-8") as f:
        for image_name, text in results:
            f.write(f"Results for {image_name}:\n")
            if text:
                for line in text:
                    f.write(line + "\n")
            else:
                f.write("No text detected.\n")
            f.write("\n")

    print("OCR extraction completed. Results saved in", output_file)

if __name__ == "__main__":
    main()
```

```python
import os

def extract_password_lines(input_file, output_file):
    # Check if the input file exists
    if not os.path.exists(input_file):
        print(f"Error: {input_file} does not exist.")
        return

    # Open the input file and read all lines
    with open(input_file, "r", encoding="utf-8") as infile:
        lines = infile.readlines()

    # Filter for lines that contain the word "Password"
    password_lines = [line for line in lines if "Password" in line]

    # Write the filtered lines to the output file
    with open(output_file, "w", encoding="utf-8") as outfile:
        outfile.writelines(password_lines)

    print(f"Copied {len(password_lines)} line(s) containing 'Password' from {input_file} to {output_file}.")

if __name__ == "__main__":
    # Specify your OCR output file and the new file for password lines
    input_file = "output.txt"          # Replace with your OCR output file name if different
    output_file = "password_lines.txt" # This file will contain the lines with "Password"

    extract_password_lines(input_file, output_file)
```
